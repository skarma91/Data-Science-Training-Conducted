## Matplotlib

- Matplotlib is an open source python library which is used for plotting the data and providing data visualization.
- There is one particular submodule that we will mostly use: matplotlib.pyplot
- If one is using anaconda distribution then matplotlib comes pre-installed.
- If matplotlib is not installed in the system one can easily install it using the command `pip install matplotlib`

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

## Line Plots

```python
x = np.linspace(0, 4*np.pi, 1000)  # create 1000 equally spaced datapoints between 0 to 4*pi 
```

```python
y = np.sin(x)  
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, y)
plt.title("sin(x) vs x", fontsize=18)
plt.xlabel("x", fontsize=12)
plt.ylabel("sin(x)", fontsize=12)
plt.grid()
plt.show()
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, y, label='sin(x)')
plt.plot(x, z, 'r--', label='cos(x)')
plt.title("sin(x) and cos(x) vs x", fontsize=18)
plt.xlabel("x", fontsize=12)
plt.ylabel("sin(x) / cos(x)", fontsize=12)
plt.grid()
plt.legend(fontsize=12)
plt.show()
```

```python
attendance = [60, 72, 55, 50, 75, 85, 90, 71, 53, 92]
marks = [73, 81, 60, 48, 82, 91, 89, 76, 82, 99]
```

```python
height_weight_sample.groupby('Gender').count()
```

```python
height_weight_sample_male = height_weight_sample[height_weight_sample['Gender']=='Male']
height_weight_sample_female = height_weight_sample[height_weight_sample['Gender']=='Female']
```

```python
plt.figure(figsize=(12,8))
plt.scatter(height_weight_sample_male['Height'], height_weight_sample_male['Weight'], s=30, edgecolors='k', label='male', c='m')
plt.scatter(height_weight_sample_female['Height'], height_weight_sample_female['Weight'], s=30, edgecolors='k', label='female', c='c')
plt.title("Weight vs. Height of adults", fontsize=18)
plt.xlabel("Heights", fontsize=14)
plt.ylabel("Weights", fontsize=14)
plt.grid()
plt.legend(fontsize=14)
plt.show()
```

## Bar Plot

```python
medals_tally = {
    'country': ['Aus', 'Eng', 'Can', 'Ind', 'NewZ'],
    'gold': [67, 57, 26, 22, 20],
    'silver': [57, 66, 32, 16, 12],
    'bronze': [54, 53, 34, 23, 17],
    'total': [178, 176, 92, 61, 49],
}
```

```python
medals_cw_2022 = pd.DataFrame(medals_tally).sort_values('country').reset_index(drop=True)
```

```python
medals_cw_2022
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=medals_cw_2022['country'], height=medals_cw_2022['total'])
plt.title("CW-2022 medals by different countries", fontsize=18)
plt.ylabel("Total Medals")
```

### Stacked bar plot

```python
bronze = medals_cw_2022['bronze']
silver = medals_cw_2022['silver']
gold = medals_cw_2022['gold']
countries = medals_cw_2022['country']
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=countries, height=bronze, label='bronze', color='g',width=0.5)
plt.bar(x=countries, height=silver, bottom=bronze, label='silver', color= 'b',width=0.5)
plt.bar(x=countries, height=gold, bottom=bronze+silver, label='gold', color= 'r',width=0.5)
plt.title("CW-2022 medals by different countries", fontsize=18)
plt.ylabel("Total Medals", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

### Grouped bar chart

```python
x = np.arange(len(medals_cw_2022))
w = 0.2
plt.figure(figsize=(12,8))
plt.bar(x-w, height=bronze, label='bronze', color='c', width=w)
plt.bar(x, height=silver, label='silver', color= 'm', width=w)
plt.bar(x+w, height=gold, label='gold', color= 'b', width=w)
plt.xticks(x, countries)
plt.title("CW-2022 medals by different countries", fontsize=18)
plt.ylabel("Total Medals", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

## pie-plot

```python
plt.figure(figsize=(8,8))
plt.pie(x=medals_cw_2022['total'], labels=countries, autopct="%1.2f%%")
plt.show()
```

## subplots

```python
x = np.linspace(0, 4*np.pi, 1000)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
y4 = np.tan(x)
```

```python
figure, axis = plt.subplots(2,2)

axis[0, 0].plot(x, y1)
axis[0, 0].set_title('sin(x) vs x')

axis[0, 1].plot(x, y2)
axis[0, 1].set_title('cos(x) vs x')

axis[1, 0].plot(x, y3)
axis[1, 0].set_title('sin(x)*cos(x) vs x')

axis[1, 1].plot(x, y4)
axis[1, 1 ].set_title('tan(x) vs x')

plt.suptitle("Plot of various functions")
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.suptitle("Plot of various functions", fontsize=15)

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("sin(x)",fontsize=12)

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('cos(x)',fontsize=12)

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('sin(x)*cos(x)',fontsize=12)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('tan(x)',fontsize=12);
```
