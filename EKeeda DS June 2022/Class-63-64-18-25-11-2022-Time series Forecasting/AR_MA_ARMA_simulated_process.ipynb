## Simulated AR process

An Auto-regressive process of order $p$ called $AR(p)$ can be described as:

$$X_t = \sum_{i=1}^{p}\phi_i X_{t-i} + \epsilon_t $$

Here, $\epsilon_t$ is called the white noise will be modelled as normal distribution with zero mean and variance $\sigma^2$

For simplicity we shall consider AR process of order 1. 

i.e. we shall simulate 

$$X_t = \phi_1 X_{t-1} + \epsilon_t$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulated_AR(p=1, phi=[1], x0=1.0, sigma=1.0, n=100):
    """
    This function simulates AR process of order p.
    
    inputs:
      - p: int -> order of the AR process (default 1)
      - phi: list of ints -> Lag coefficients (default [1])
      - x0: float -> initial value of time series, also called initial condition (default 1.0)
      - sigma: float -> the standard deviation of noise
      - n: int -> number of datapoints to be generated
    
    outputs:
     - list of n values generated by AR process described by the parameters given as input
    """
    
    x_out = []
    x_prev = x0
    for i in range(n):
        epsilon = sigma*(np.random.randn(1)[0])
        x = phi[0]*x_prev + epsilon
        x_out.append(x)
    return x_out
```

```python
X = simulated_AR(phi=[1.5], x0=0, sigma=2)
```

```python
plt.figure(figsize=(12,6))
plt.plot(X, marker='o')
plt.grid()
plt.title(f"Simulated AR of order 1")
plt.show()
```
