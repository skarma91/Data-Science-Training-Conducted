## Examine the relationship between height and weight (Q -> Q)

```python
sample_data = data.sample(frac = 0.2)
```

```python
plt.figure(figsize=(12,9))
seaborn.scatterplot(x=sample_data['Height'], y=sample_data['Weight'], hue=sample_data['Gender'])
plt.legend()
plt.title("Scatter plot of Weights vs Heights", fontsize = 14)
plt.show()
```

```python
def covariance(X, Y):
    x_bar = X.mean()
    y_bar = Y.mean()
    cov = ((X - x_bar) @ (Y - y_bar)) / X.shape[0]
    return cov
```

```python
covariance(sample_data['Height'].values, sample_data['Weight'].values)
```

```python
def correlation(X, Y):
    cov = covariance(X, Y)
    std_x = X.std()
    std_y = Y.std()
    corr = cov / (std_x * std_y)
    return corr
```

```python
correlation(sample_data['Height'].values, sample_data['Weight'].values)
```
